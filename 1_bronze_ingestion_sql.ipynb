```sql
DROP SCHEMA IF EXISTS workspace.retail_etl CASCADE;
     

CREATE SCHEMA IF NOT EXISTS training_catalog.bronze;
USE training_catalog.bronze;
     

DROP TABLE IF EXISTS training_catalog.bronze.customers_raw;
DROP TABLE IF EXISTS training_catalog.bronze.products_raw;
DROP TABLE IF EXISTS training_catalog.bronze.stores_raw;
DROP TABLE IF EXISTS training_catalog.bronze.sales_raw;
     

SHOW TABLES IN training_catalog.bronze;
     

CREATE SCHEMA IF NOT EXISTS training_catalog.bronze;
USE training_catalog.bronze;

CREATE TABLE IF NOT EXISTS customers_raw
USING CSV
OPTIONS (
  header = 'true',
  inferSchema = 'true'
)
LOCATION 's3a://sneha-retail-sales-data-warehouse/sftp/customers/';
     

CREATE TABLE IF NOT EXISTS products_raw
USING CSV
OPTIONS (
  header = 'true',
  inferSchema = 'true'
)
LOCATION 's3a://sneha-retail-sales-data-warehouse/sftp/products/';
     

CREATE TABLE IF NOT EXISTS stores_raw
USING CSV
OPTIONS (
  header = 'true',
  inferSchema = 'true'
)
LOCATION 's3a://sneha-retail-sales-data-warehouse/sftp/stores/';
     

CREATE TABLE IF NOT EXISTS sales_raw
USING CSV
OPTIONS (
  header = 'true',
  inferSchema = 'true'
)
LOCATION 's3a://sneha-retail-sales-data-warehouse/sftp/sales/';
     

SELECT 'customers' AS table_name, COUNT(*) AS row_count FROM customers_raw
UNION ALL
SELECT 'products',  COUNT(*) FROM products_raw
UNION ALL
SELECT 'stores',    COUNT(*) FROM stores_raw
UNION ALL
SELECT 'sales',     COUNT(*) FROM sales_raw;
     
```
